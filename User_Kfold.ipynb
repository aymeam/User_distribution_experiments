```python
data = pd.read_csv('DatosCSV\en_total.csv','r',delimiter=',')
import numpy as np
X=[]#usuarios
y=[]#clase
for line in data.values:
    y.append(line[0]) 
    X.append(line[3])
X=np.array(X)
y=np.array(y)
```
